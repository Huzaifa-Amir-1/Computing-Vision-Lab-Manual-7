```python
# ==========================================
# INSTALL REQUIRED PACKAGES
# ==========================================
!pip install opencv-python-headless scikit-learn --quiet


# ==========================================
# IMPORT LIBRARIES
# ==========================================
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
from google.colab import files
import os
import zipfile


# ==========================================
# GABOR FEATURE EXTRACTION FUNCTION
# ==========================================
def gabor_features(img, kernels):
    features = []

    for kernel in kernels:
        filtered = cv2.filter2D(img, cv2.CV_8UC3, kernel)

        mean = filtered.mean()
        variance = filtered.var()

        features.extend([mean, variance])

    return features


# ==========================================
# STEP 1: UPLOAD ZIP DATASET
# ==========================================
print("Upload your texture_dataset.zip file")
uploaded = files.upload()

zip_filename = list(uploaded.keys())[0]


# ==========================================
# STEP 2: EXTRACT ZIP
# ==========================================
with zipfile.ZipFile(zip_filename, 'r') as zip_ref:
    zip_ref.extractall('texture_dataset')


# Detect actual dataset folder automatically
dataset_path = 'texture_dataset'

inside = os.listdir(dataset_path)

if len(inside) == 1 and os.path.isdir(os.path.join(dataset_path, inside[0])):
    dataset_path = os.path.join(dataset_path, inside[0])


print("Dataset Loaded From:", dataset_path)


# ==========================================
# STEP 3: CREATE GABOR KERNELS
# ==========================================
kernel_size = 21

sigmas = [4.0, 6.0]
lambdas = [10.0, 15.0]
gammas = [0.5]
psis = [0]
orientations = [0, np.pi/4, np.pi/2, 3*np.pi/4]

kernels = []

for sigma in sigmas:
    for lambd in lambdas:
        for gamma in gammas:
            for psi in psis:
                for theta in orientations:

                    kernel = cv2.getGaborKernel(
                        (kernel_size, kernel_size),
                        sigma,
                        theta,
                        lambd,
                        gamma,
                        psi,
                        ktype=cv2.CV_32F
                    )

                    kernels.append(kernel)


print("Total Gabor Kernels Created:", len(kernels))


# ==========================================
# STEP 4: LOAD IMAGES + EXTRACT FEATURES
# ==========================================
X = []
y = []

label_map = {}
label_id = 0


for class_name in os.listdir(dataset_path):

    class_path = os.path.join(dataset_path, class_name)

    if os.path.isdir(class_path):

        label_map[label_id] = class_name

        for img_name in os.listdir(class_path):

            img_path = os.path.join(class_path, img_name)

            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

            if img is not None:

                # Resize all images to fixed size
                img = cv2.resize(img, (128, 128))

                features = gabor_features(img, kernels)

                X.append(features)

                y.append(label_id)

        label_id += 1


# Convert to numpy arrays
X = np.array(X)
y = np.array(y)


# Safety check
if len(X) == 0:
    raise Exception("No images found! Check your ZIP structure.")


print("Total Images Loaded:", len(X))
print("Classes Found:", label_map)


# ==========================================
# STEP 5: TRAIN TEST SPLIT
# ==========================================
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)


# ==========================================
# STEP 6: TRAIN SVM MODEL
# ==========================================
clf = SVC(kernel='linear')

clf.fit(X_train, y_train)


# ==========================================
# STEP 7: PREDICT
# ==========================================
y_pred = clf.predict(X_test)


# ==========================================
# STEP 8: EVALUATION
# ==========================================
print("\n===== CLASSIFICATION REPORT =====\n")

print(classification_report(
    y_test,
    y_pred,
    target_names=[label_map[i] for i in sorted(label_map.keys())]
))

accuracy = accuracy_score(y_test, y_pred)

print(f"\nOverall Accuracy: {accuracy*100:.2f}%")


# ==========================================
# STEP 9: VISUALIZE PREDICTIONS
# ==========================================
plt.figure(figsize=(12, 6))

for i in range(min(6, len(y_pred))):

    plt.subplot(2, 3, i+1)

    plt.text(
        0.5,
        0.5,
        f"Predicted:\n{label_map[y_pred[i]]}",
        fontsize=12,
        ha='center',
        va='center'
    )

    plt.axis('off')


plt.suptitle("Sample Predicted Texture Classes")

plt.tight_layout()

plt.show()
```

```python
import cv2
import numpy as np
import os
import zipfile
from google.colab import files
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

print("Upload Tile Dataset ZIP:")
uploaded = files.upload()

zip_name = list(uploaded.keys())[0]

with zipfile.ZipFile(zip_name,'r') as zip_ref:
    zip_ref.extractall("tile_dataset")

def gabor_features(img):
    kernel = cv2.getGaborKernel((21,21),5,np.pi/4,10,0.5,0)
    filtered = cv2.filter2D(img,cv2.CV_8UC3,kernel)
    return [filtered.mean(), filtered.var()]

X=[]
y=[]

for label,folder in enumerate(os.listdir("tile_dataset")):
    for file in os.listdir(f"tile_dataset/{folder}"):

        img=cv2.imread(f"tile_dataset/{folder}/{file}",0)

        X.append(gabor_features(img))
        y.append(label)

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)

model=SVC()

model.fit(X_train,y_train)

print("Accuracy:",model.score(X_test,y_test))
```

```python
import cv2
import numpy as np
import os
import zipfile
from google.colab import files
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

print("Upload Material Dataset ZIP:")
uploaded = files.upload()

zip_name = list(uploaded.keys())[0]

with zipfile.ZipFile(zip_name,'r') as zip_ref:
    zip_ref.extractall("material_dataset")

def extract_features(img):

    kernel=cv2.getGaborKernel((21,21),6,np.pi/4,12,0.5,0)

    filtered=cv2.filter2D(img,cv2.CV_8UC3,kernel)

    return [filtered.mean(),filtered.var()]

X=[]
y=[]

for label,folder in enumerate(os.listdir("material_dataset")):

    for file in os.listdir(f"material_dataset/{folder}"):

        img=cv2.imread(f"material_dataset/{folder}/{file}",0)

        X.append(extract_features(img))
        y.append(label)

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)

clf=RandomForestClassifier()

clf.fit(X_train,y_train)

print("Accuracy:",clf.score(X_test,y_test))
```
